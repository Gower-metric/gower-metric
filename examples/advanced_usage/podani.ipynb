# Podani + conditional distances with custom threshold

```python
import numpy as np

from gower_metric import Gower
from gower_metric.core.config import Config
```

```python
DTYPE = np.float64

data = np.array(
    [[1, "low", 3.5], [2, "medium", 4.0], [3, "high", 2.5], [4, "medium", 5.0]],
    dtype=object,
)

feature_types = {
    0: "ratio_scale_interval",
    1: "categorical_ordinal",
    2: "ratio_scale_interval",
}

ordinal_mappings = {1: ["low", "medium", "high"]}

cfg = Config(
    feature_types=feature_types,
    data_type=DTYPE,
    categorical_ordinal_values_order=ordinal_mappings,
    categorical_ordinal_calculation_type="podani",
    conditional_distances=True,  # <--- please remember to set this to True if using conditional distances threshold, otherwise it will lead to unexpected behavior
    conditional_distances_threshold_coeff=3,
)
gower = Gower(cfg).fit(data)
```

```python
result = gower(data[0], data[1])
```
