```python
import numpy as np

from gower_metric import Config, Gower
```

```python
DTYPE = np.float16

data = np.array(
    [[1, "a", 3.5], [2, "b", 4.0], [3, "a", 2.5], [4, "c", 5.0]],
    dtype=object,
)

feature_types = {
    0: "ratio_scale_interval",
    1: "categorical_nominal",
    2: "ratio_scale_interval",
}

weights = {0: 0.5, 1: 2.0, 2: 1.0}

cfg = Config(
    feature_types=feature_types,
    feature_weights=weights,
    data_type=DTYPE,
)
gower = Gower(cfg).fit(data)
```

```python
result = gower(data[0], data[1])
```
