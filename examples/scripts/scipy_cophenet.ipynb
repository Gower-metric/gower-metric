```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import single, cophenet

from gower_metric import Gower
```

```python
n_rows = 100
df = pd.read_csv('../../comparison/data/adult.csv').head(n_rows)

df = df[['age', 'educational-num', 'race', 'gender', 'hours-per-week', 'relationship', 'occupation', 'education', 'workclass']]
df.replace(to_replace='?', value=np.nan, inplace=True)
```

```python
print(df.head())
```

```python
feature_types = {
    "age": "ratio_scale_interval",
    "educational-num": "ratio_scale_interval",
    "race": "categorical_nominal",
    "gender": "categorical_nominal",
    "hours-per-week": "ratio_scale_interval",
    "relationship": "categorical_nominal",
    "occupation": "categorical_nominal",
    "education": "categorical_nominal",
    "workclass": "categorical_nominal"
}
```

```python
gower = Gower(feature_types=feature_types).fit(df)
df = df.to_numpy()
```

```python
Z = single(pdist(df, metric=gower))
c, _ = cophenet(Z, pdist(df, metric=gower))
```

```python
print(f"Cophenetic correlation coefficient: {c:.4f}")
```

```python
print(f"{squareform(cophenet(Z))}")
```
