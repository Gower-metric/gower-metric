```python
import warnings

import numpy as np
import pandas as pd
from sklearn.metrics import pairwise_distances

from gower_metric import Gower

warnings.filterwarnings("ignore")
```

There is also possibility to use scikit-learn pairwaise distances with our API. Because one will also operate on string and boolean values, we have to pass custom function as well.

```python
n_rows = 100
df = pd.read_csv("../../comparison/data/adult.csv").head(n_rows)

df = df[
    [
        "age",
        "educational-num",
        "race",
        "gender",
        "hours-per-week",
        "relationship",
        "occupation",
        "education",
        "workclass",
    ]
]
```

```python
# process data -> where spot '?' replace with NaN
df.replace(to_replace="?", value=np.nan)
```

```python
print(df.head())
```

```python
feature_types = {
    "age": "ratio_scale_interval",
    "educational-num": "ratio_scale_interval",
    "race": "categorical_nominal",
    "gender": "categorical_nominal",
    "hours-per-week": "ratio_scale_interval",
    "relationship": "categorical_nominal",
    "occupation": "categorical_nominal",
    "education": "categorical_nominal",
    "workclass": "categorical_nominal",
}
```

```python
gower = Gower(feature_types=feature_types).fit(df)
```

```python
matrix = pairwise_distances(df, metric=gower, n_jobs=-1, ensure_all_finite=False)
```

```python
print("Gowers distance matrix: \n", matrix)
```
