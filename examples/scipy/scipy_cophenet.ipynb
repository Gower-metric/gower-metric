```python
import warnings

import numpy as np
import pandas as pd

from gower_metric import Gower
from gower_metric.core.config import Config
from scipy.cluster.hierarchy import cophenet, single
from scipy.spatial.distance import pdist, squareform

warnings.filterwarnings("ignore")
```

```python
n_rows = 1_000
df = pd.read_csv("../../data/files/adult.csv").head(n_rows)

df = df[
    [
        "age",
        "educational-num",
        "race",
        "gender",
        "hours-per-week",
        "relationship",
        "occupation",
        "education",
        "workclass",
    ]
]
df.replace(to_replace="?", value=np.nan)
```

```python
feature_types = {
    "age": "ratio_scale_interval",
    "educational-num": "ratio_scale_interval",
    "race": "categorical_nominal",
    "gender": "categorical_nominal",
    "hours-per-week": "ratio_scale_interval",
    "relationship": "categorical_nominal",
    "occupation": "categorical_nominal",
    "education": "categorical_nominal",
    "workclass": "categorical_nominal",
}
```

```python
cfg = Config(
    feature_types=feature_types,
)
gower = Gower(cfg).fit(df)
df_num = gower.transform(df)
```

```python
Z = single(pdist(df_num, metric=gower))
c, _ = cophenet(Z, pdist(df_num, metric=gower))
```

```python
print(f"Cophenetic correlation coefficient: {c:.4f}")
```

```python
print(f"{squareform(cophenet(Z))}")
```
