```python
import warnings

import numpy as np
import pandas as pd
from sklearn.metrics import pairwise_distances

from gower_metric import Gower

warnings.filterwarnings("ignore")
```

```python
n_rows = 1_000
df = pd.read_csv("../../data/files/adult.csv").head(n_rows)

df = df[
    [
        "age",
        "educational-num",
        "race",
        "gender",
        "hours-per-week",
        "relationship",
        "occupation",
        "education",
        "workclass",
    ]
]
```

```python
df.replace(to_replace="?", value=np.nan, inplace=True)
```

```python
feature_types = {
    "age": "ratio_scale_interval",
    "educational-num": "ratio_scale_interval",
    "race": "categorical_nominal",
    "gender": "categorical_nominal",
    "hours-per-week": "ratio_scale_interval",
    "relationship": "categorical_nominal",
    "occupation": "categorical_nominal",
    "education": "categorical_nominal",
    "workclass": "categorical_nominal",
}
```

```python
gower = Gower(feature_types=feature_types).fit(df)
df_num = gower.transform(df)
```

```python
matrix_scikit = pairwise_distances(df_num, metric=gower, n_jobs=-1, ensure_all_finite=False)
```

```python
matrix_gower = gower.matrix(df)
```

```python
assert np.allclose(matrix_scikit, matrix_gower, atol=1e-5), "matrices are not equal"
```
