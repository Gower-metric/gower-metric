```python
import numpy as np
import pandas as pd

from gower_metric import Gower
from gower_metric.core.config import Config
```

```python
iris = pd.read_csv("../../../data/files/iris.csv")

f_types: dict[int | str, str] = {
    "sepal_length": "numeric",
    "sepal_width": "numeric",
    "petal_length": "numeric",
    "petal_width": "numeric",
    "variety": "categorical_nominal",
}
cfg = Config(
    feature_types=f_types,
    scale_method="range",
)
gower = Gower(cfg).fit(iris)

rows_1 = range(10)
rows_2 = range(5, 15)

py_block = np.zeros((10, 10))
for i, r1 in enumerate(rows_1):
    for j, r2 in enumerate(rows_2):
        py_block[i, j] = gower(iris.iloc[r1], iris.iloc[r2])
```

```python
np.savetxt("daisy_results/daisy_python_results.txt", py_block, fmt="%.6f")
```
