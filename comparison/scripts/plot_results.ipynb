```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cran_python_gower_speed = np.loadtxt("cran/cran_results/speed_cran_gower_python.txt")
cran_r_speed = np.loadtxt("cran/cran_results/speed_cran_gower.txt")

daisy_python_speed = np.loadtxt("daisy/daisy_results/daisy_python_speed.txt")
daisy_r_speed = np.loadtxt("daisy/daisy_results/daisy_speed.txt")

means_cran = [np.mean(cran_r_speed), np.mean(cran_python_gower_speed)]
stds_cran = [np.std(cran_r_speed), np.std(cran_python_gower_speed)]

means_daisy = [np.mean(daisy_r_speed), np.mean(daisy_python_speed)]
stds_daisy = [np.std(daisy_r_speed), np.std(daisy_python_speed)]

N_ITERATIONS_cran = len(cran_python_gower_speed)
N_ITERATIONS_daisy = len(daisy_python_speed)
```

```python
def plot_cran_gower_speed_comparison(
    means=means_cran,
    stds=stds_cran,
    N_ITERATIONS=N_ITERATIONS_cran,
) -> None:
    """Plot speed comparison between R and Python implementations of Gower similarity from CRAN.

    Args:
        means (list): Mean execution times for R and Python implementations.
        stds (list): Standard deviations of execution times for R and Python implementations.
        N_ITERATIONS (int): Number of iterations used for timing.

    """
    plt.figure(figsize=(10, 6))
    x_pos = [0, 1]

    plt.bar(
        x_pos,
        means,
        yerr=stds,
        capsize=10,
        alpha=0.7,
        color="darkblue",
        edgecolor="black",
        linewidth=1.5,
    )

    plt.xticks(x_pos, ["R", "Python"])
    plt.ylabel("Execution Time (seconds)")
    plt.title(f"Speed comparison over {N_ITERATIONS} Iterations (Mean ± Std Dev)")
    plt.grid(axis="y", alpha=0.3)

    for i, (mean, std) in enumerate(zip(means, stds, strict=False)):
        plt.text(
            i,
            mean + std + 0.01,
            f"{mean:.4f}±{std:.4f}",
            ha="center",
            va="bottom",
        )

        plt.tight_layout()
        plt.savefig("../../data/imgs/comparison/cran_gower_similarity_comparison.svg")


plot_cran_gower_speed_comparison()
```

```python
def plot_daisy_speed_comparison(
    means=means_daisy,
    stds=stds_daisy,
    N_ITERATIONS=N_ITERATIONS_daisy,
) -> None:
    """Plot speed comparison between R and Python implementations of daisy.

    Args:
        means (list, optional): Mean execution times for R and Python. Defaults to means_daisy.
        stds (list, optional): Standard deviations of execution times for R and Python. Defaults to stds_daisy.
        N_ITERATIONS (int, optional): Number of iterations used in the comparison. Defaults to N_ITERATIONS_daisy.

    """
    plt.figure(figsize=(10, 6))
    x_pos = [0, 1]

    plt.bar(
        x_pos,
        means,
        yerr=stds,
        capsize=10,
        alpha=0.7,
        color="darkblue",
        edgecolor="black",
        linewidth=1.5,
    )

    plt.xticks(x_pos, ["R", "Python"])
    plt.ylabel("Execution Time (seconds)")
    plt.title(f"Speed comparison over {N_ITERATIONS} Iterations (Mean ± Std Dev)")
    plt.grid(axis="y", alpha=0.3)

    for i, (mean, std) in enumerate(zip(means, stds, strict=False)):
        plt.text(
            i,
            mean + std + 0.01,
            f"{mean:.4f}±{std:.4f}",
            ha="center",
            va="bottom",
        )

        plt.tight_layout()
        plt.savefig("../../data/imgs/comparison/daisy_gower_distance_comparison.svg")


plot_daisy_speed_comparison()
```
